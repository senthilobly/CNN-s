```python
# CNN network with padding
import numpy as np
import matplotlib.pyplot as plt
input = np.random.randint(1,10,(28,28))
kernel = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
padding = (kernel.shape[0] - 1)//2
stride = 1
feature_map = ((input.shape[0] - kernel.shape[0]) + (2 * padding) // 1) + 1
padded_input = np.pad(input,(padding, padding), mode='constant', constant_values=0)
output = np.zeros((feature_map, feature_map))
print("\nInput size before Padding: {}".format(input.shape))
print("\nInput size after padding: {}".format(padded_input.shape))
for h in range(feature_map):
  for w in range(feature_map):
    region = padded_input[h:h+kernel.shape[0], w:w+kernel.shape[1]]
    output[h,w] = np.sum(region * kernel)
print("\nConvolved output:")
print("\n",output)
```
