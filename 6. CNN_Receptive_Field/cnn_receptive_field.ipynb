```python
# Basic CNN program for Receptive Field understanding
import numpy as np
import matplotlib.pyplot as plt
input = np.random.randint(1,10,(28,28))
kernel_in = np.ones((3,3))
stride_in = 1

def conv_2d(image, kernel, stride=1):
  ker_h, ker_w = kernel.shape
  out_h = ((image.shape[0]-ker_h)//stride)+1
  out_w = ((image.shape[1]-ker_w)//stride)+1
  output = np.zeros((out_h, out_w))
  for i in range(out_h):
    for j in range(out_w):
      region = image[i*stride:i*stride+ker_h, j*stride:j*stride+ker_w]
      output[i,j] = np.sum(region * kernel)
  return output

conv1 = conv_2d(input, kernel_in, stride=stride_in)
print("\nLayer 1 output shape:", conv1.shape)
conv2 = conv_2d(conv1, kernel_in, stride=2)
print("\nLayer 2 output shape:", conv2.shape)
conv3 = conv_2d(conv2, kernel_in, stride=2)
print("\nLayer 3 output shape:", conv3.shape)

plt.figure(figsize=(12,5))
plt.subplot(1,4,1)
plt.imshow(input)
plt.title("Input Image")

plt.subplot(1,4,2)
plt.imshow(conv1, cmap="BrBG_r")
plt.title("CNN Layer 1")

plt.subplot(1,4,3)
plt.imshow(conv2, cmap="CMRmap_r")
plt.title("CNN Layer 2")

plt.subplot(1,4,4)
plt.imshow(conv3, cmap="RdYlGn")
plt.title("CNN Layer 3")

plt.show()
```
